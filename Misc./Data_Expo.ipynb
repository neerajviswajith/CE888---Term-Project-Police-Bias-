```
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
ar_df = pd.read_csv('/content/drive/MyDrive/CE888 - Project/Arrests_data.csv')
ar_df.head(5)
```

```
ar_df.columns
```

```
# Data Pre-pr0cessing
ar_df.drop(columns=['Measure','Time_type', 'Ethnicity_type','Gender_type',
                    'Age_Group_type','Geography_type'], inplace=True)
```

```
# Statistis on number of yearly arrests from 2006/07 - 2017/18
yearly_arrests = ar_df.loc[ar_df['Ethnicity'] == 'All'].loc[ar_df['Gender'] == 'All'].loc[ar_df['Age_Group'] == 'All'].loc[ar_df['Geography'] == 'All']
yearly_arrests = yearly_arrests[['Time',' Number of arrests ']]
yearly_arrests[' Number of arrests '] = [x.replace(',','') for x in yearly_arrests[' Number of arrests ']]
yearly_arrests[' Number of arrests '] = yearly_arrests[' Number of arrests '].astype(int)

#Plotting Values

plt.figure(figsize=(15,7.5),facecolor='w')
plt.bar(yearly_arrests['Time'], yearly_arrests[' Number of arrests '], color='grey')
plt.xlabel("Year")
plt.ylabel("Number Of Arrests")
plt.title("Total Arrest per year 2007-18")
plt.show()
```

```
geo_ar = ar_df.loc[ar_df['Ethnicity'] == 'All'].loc[ar_df['Gender'] == 'All'].loc[ar_df['Age_Group'] == 'All'].loc[ar_df['Geography'] != 'All']
geo_ar = geo_ar[['Time','Geography',' Number of arrests ']]
geo_ar = geo_ar.loc[geo_ar['Geography'] != 'Lancashire']
geo_ar[' Number of arrests '] = [x.replace(',','') for x in geo_ar[' Number of arrests ']]
geo_ar[' Number of arrests '] = geo_ar[' Number of arrests '].astype(int)
geo_ar = geo_ar.groupby(geo_ar['Geography']).mean()
geo_ar.sort_values(' Number of arrests ', ascending=False)
```
