## Importing Libraries and Datasets

```python
# Importing Libraries

import pandas as pd
import numpy as np
```

## Stop and Search Data 2011
###### This data has been choosen to match the ethnic population data retrieved from Cesus 2011

```python
# Fetching 2011 Stop and Search Data for Analysis
# Source: https://www.gov.uk/government/statistics/tables-police-powers-and-procedures-england-and-wales-2010-to-2011

!wget 'https://assets.publishing.service.gov.uk/government/uploads/system/uploads/attachment_data/file/115724/stops-searches-1011-tabs.xls'
```

###### SS.12 - Number of persons and vehicles searched under section 1 of the Police and Criminal Evidence Act 1984 By Ethnicity


```python
SS1 = pd.read_excel('stops-searches-1011-tabs.xls', sheet_name='SS.12', header=5)
SS1.head(5)
```

```python
# Dropping uncessary columns
# Here, the Not stated columsn is dropped as ethinicity unclear

SS1.drop(columns=['Vehicle only search','Total','Not \nstated'], inplace=True)
SS1.dropna(inplace=True)
SS1.columns = ['Geography','SS1_White','SS1_Black','SS1_Asian','SS1_Other','SS1_Mixed']
SS1.reset_index(inplace=True, drop=True)
SS1.head(5)
```

###### SS.14 - Number of persons and vehicles searched under section 60 of the Criminal Justice and Public Order Act 1994, by Ethnicity

```python
SS2 = pd.read_excel('stops-searches-1011-tabs.xls', sheet_name='SS.14', header=5)
SS2.head(5)
```

```python
# Dropping uncessary columns
# Here, the Not stated columsn is dropped as ethinicity unclear

SS2.drop(columns=['Vehicle only search','Total','Not \nstated'], inplace=True)
SS2.dropna(inplace=True)
SS2.columns = ['Geography','SS2_White','SS2_Black','SS2_Asian','SS2_Other','SS2_Mixed']
SS2.reset_index(inplace=True, drop=True)
SS2.head(5)
```

###### SS.16 - Number of persons and vehicles searched under section 60 of the Criminal Justice and Public Order Act 1994, by


```python
SS3 = pd.read_excel('stops-searches-1011-tabs.xls', sheet_name='SS.16', header=5)
SS3.head(5)
```

```python
# Dropping uncessary columns
# Here, the Not stated columsn is dropped as ethinicity unclear

SS3 = SS3.iloc[:68,:7]
SS3 = SS3.drop(columns='Unnamed: 1')
SS3.columns = ['Geography','SS3_White','SS3_Black','SS3_Asian','SS3_Other','SS3_Mixed']
SS3.dropna(inplace=True)
SS3.reset_index(inplace=True, drop=True)
SS3.head(5)
```

```python
SS = SS1.merge(SS2.merge(SS3, on='Geography'), on='Geography')
sum = SS.iloc[:,6:11].values + SS.iloc[:,1:6].values + SS.iloc[:,11:].values
stop = pd.DataFrame(sum, columns=['SS_White','SS_Black','SS_Asian','SS_Other','SS_Mixed'])
stop = pd.concat([SS['Geography'],stop], axis=1)
stop.head(5)
```

```python
stop.to_csv('ss_data.csv')
```

## Arrest Data 2011

```python
# Fetching 2011 Arrest Data for Analysis
# Source: https://www.gov.uk/government/statistics/tables-police-powers-and-procedures-england-and-wales-2010-to-2011

!wget 'https://assets.publishing.service.gov.uk/government/uploads/system/uploads/attachment_data/file/115722/arrests-1011-tabs.xls'
```

```python
# Data load and Pre-processing

AR = pd.read_excel('arrests-1011-tabs.xls', sheet_name='A.11', header=7)
AR = pd.concat([AR.iloc[:,0],AR.iloc[:67,18:]], axis=1) 
AR = AR.iloc[:,[0,1,3,5,7,9]]
AR = AR.dropna()
AR = AR.reset_index(drop=True)
AR.columns = ['Geography','AR_White','AR_Black','AR_Asian','AR_Mixed','AR_Other']
AR = AR[['Geography','AR_White','AR_Black','AR_Asian','AR_Other','AR_Mixed']]
AR.head(5)
```

## Creating Master DataFarme for Further Calculation

##### Merging Stop & Search, Arrests and Population Data to Create Master Datatset

```python
# Retrieving popultion data from project Repo

pop = pd.read_csv('https://raw.githubusercontent.com/neerajthandayan/CourseProject/main/Data/pop_data.csv')
pop.drop(columns='Unnamed: 0', inplace=True)
pop.head(5)
```

```python
# Merging dataset

master = pop.merge(stop.merge(AR, on='Geography'), on='Geography')
master
```

## Calculating Search and Stop Rate by Ethinicity

```python
rate = pd.DataFrame((master.iloc[:,6:11].values / master.iloc[:,1:6].values)*100, columns= master.iloc[:,6:11].columns)
ssrate = pd.concat([master['Geography'],rate], axis=1)
ssrate.head()
```

```python
ssrate.to_csv('ss_rate.csv')
```

## Calculating Arrest Rate by Ethinicity

```python
rt = pd.DataFrame((master.iloc[:,11:].values / master.iloc[:,1:6].values), columns= master.iloc[:,11:].columns)
arate = pd.concat([master['Geography'],rt.astype(float,errors='raise') ], axis=1)
for i in arate.columns[1:]:
  arate[i] = arate[i].copy() * ssrate.iloc[:,1:].copy().sum(axis=1)
arate.head()
```

```python
arate.to_csv('ar_rate.csv')
```
